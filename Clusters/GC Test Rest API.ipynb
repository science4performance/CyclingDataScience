```python
"""
Created on Wed Aug 31 11:55:18 2016
imports the latest activity files and plots power curves
@author: Gavin
"""

#from glob import glob
#path = "/users/gavin/Library/GoldenCheetah"
#athletes = [i.split('/')[-2] for i in glob(path+'/*/')]


import pandas as pd
import matplotlib.pylab as plt
plt.ion()

path = "http://localhost:12021/"


athletes = pd.read_csv(path)


for athlete in ['Gavin']: #athletes.name:
    zones = pd.read_csv(path+'/'+athlete+'/zones',skipinitialspace=True)
    
    
    history = pd.read_csv(path+'//'+athlete,skipinitialspace=True)
    history.index = history.date
    
    lastActivity = history.filename.iloc[-1]
    print(lastActivity)
    activity = pd.read_csv(path+'/'+athlete+'/activity/'+lastActivity+'?format=csv',skipinitialspace=True)
    #print(activity.head())    
    meanmax = pd.read_csv(path+'/'+athlete+'/meanmax/'+lastActivity,skipinitialspace=True)
    
    bests=pd.read_csv(path+'//'+athlete+'/meanmax/bests?since=2016/01/01',skipinitialspace=True)
    bests.plot('secs','watts',logx=True,title=athlete+': Power Curve')
```
